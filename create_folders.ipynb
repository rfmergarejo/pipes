# Create the folder for every TCE project.

This code access to the TFORMs spreadsheet and create a folder named after **Project Number**. Once the *Project Number* folder is created, a **Project Name** folder is created with the corresponding folders inside.

## Gather the information

The folder is created if on the GDrive column the value is **NO**.

```python
import os

# Standard Data Frame library
import pandas as pd

# Grant access to the user....
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('TFORMs').sheet1
rows = worksheet.get_all_values()

sheet = pd.DataFrame.from_records(rows, columns = rows[0])
```

```python
missing_folder = sheet[(sheet["GDrive"] == 'NO') & (sheet["TFORM"] == "YES")]
display(missing_folder)
```

```python
# Create all the folders inside:
def create_folder():
  names = ['Timesheets', 'Site Arrows', 'Reports', 'Photos', 'Field Notes', 
           'Dig Sheets', 'Attachments']

  photos_names = ['01. Environmental', '02. Coating', '03. Corrosion deposits', 
                  '04. ILI Features', '05. Corrosion and metal loss', 
                  '06. SCC and cracking', '07. Geometry defects', 
                  '08. Manufacturing Defects', '09. Repair and Reclaimation', 
                  '10. Extra']

  for i_names in names:
    os.makedirs(i_names)

  os.chdir('Photos')
  for jj in photos_names:
    os.mkdir(jj)
      
  return print('done')

# Checking that the folder exist.... In case it doesn't it will be created.
folders_2_create = set(missing_folder["Project Number"])
for ifolder in folders_2_create:
  if os.path.isdir(ifolder):
    create_folder()
  else:
    os.mkdir(ifolder)
    os.chdir(ifolder)
    create_folder()
```
